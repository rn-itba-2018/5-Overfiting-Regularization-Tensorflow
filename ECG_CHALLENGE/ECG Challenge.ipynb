## Sobre el dataset

En el dataset provisto por la cátedra se encuentran 53127 grabaciones de latidos. Cada látido tiene su anotación especificando que tipo de latido es. Los tipos de latidos estan identificados de la siguiente forma:

- Normales, identificados con el número 0.

- Ventriculares prematuros, identificados con el número 1.

- Nodales prematuros, identificados con el número 2.

- Supraventriculares prematuros, identificados con el número 3.

El látido anotado es el que tiene el pico en el medio de la serie temporal, la misma tiene parte del latido anterior y el latidos posterior al anotado. La serie puede ser recortada de la forma que resulte conveniente para el procesamiento de los datos, es decir, si se observa que tomando n puntos (siendo $ n < tamaño /2$) a la izquierda y derecha del medio de la medición se obtiene un mejor rendimiento 

El dataset contiene 47000 latidos normales, 5206 latidos ventriculares prematuros, 119 latidos nodales prematuros y 802 latidos supraventriculares prematuros.

Se puede dividir el dataset en el set de datos de entrenamiento y set de datos de testeo de la manera que se considere mejor.
